```
list_of_2018 = []
list_of_2014 = []
list_of_difference = []
unique_countries = np.array(export[:,1],dtype=str)
unique_countries = np.unique(unique_countries)

for i in tqdm.trange(len(unique_countries)):
  if export[i,3] == 0:
    continue
  country = unique_countries[i]
  total_ex2018 = ((np.array(country) == export_2018[:,10])*export_2018[:,31]).sum()
  total_ex2014 = ((np.array(country) == export[:,1])*export[:,3]).sum()

  list_of_2018.append((unique_countries[i],total_ex2018))
  list_of_2014.append((unique_countries[i],total_ex2014))
  list_of_difference.append((unique_countries[i],(total_ex2018-total_ex2014)/total_ex2014))
```

```
unique_countries = np.array(export[:,1],dtype=str)
unique_countries = np.unique(unique_countries)

list_of_gdp_difference = []

for i in tqdm.trange(len(unique_countries)):
  if export[i,3] == 0:
    continue
  country = unique_countries[i]
  value = ((np.array(country) == gdp[:,1])*gdp[:,12]).sum()
  list_of_gdp_difference.append((country,value))
```

```
list_of_difference = pd.DataFrame(list_of_difference)
list_of_difference.to_csv('/content/drive/MyDrive/2014/difference_exp_28.12.csv')

list_of_gdp_difference = pd.DataFrame(list_of_gdp_difference)
list_of_gdp_difference.to_csv('/content/drive/MyDrive/2014/difference_gdp_28.12.csv')
```
